```python
import keras
from keras_preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
# Data Augmentation Techniques
datagen = ImageDataGenerator(rotation_range=40,
                             width_shift_range=0.2,
                             height_shift_range=0.2,
                             shear_range=0.2,
                             zoom_range=0.2,
                             horizontal_flip=True,
                             vertical_flip=True,
                             fill_mode='nearest')
```

```python
# Import images process
img = load_img('araguler.jpg')
x = img_to_array(img)
x = x.reshape((1,) + x.shape)
```

```python
# Import another image
img1 = load_img('ChetBaker.jpg')
y = img_to_array(img1)
y = y.reshape((1,) + y.shape)
```

```python
# Produce 50 different images from a image. Save .jpeg format
i = 0
for batch in datagen.flow(x, batch_size=1,
                          save_to_dir='/content/drive/MyDrive/Colab Notebooks/Data Augmentation/Augmented Data',
                          save_format='jpeg'):
  i+=1
  if i>50:
    break
```

```python
i = 0
for batch in datagen.flow(y, batch_size=1,
                          save_to_dir='/content/drive/MyDrive/Colab Notebooks/Data Augmentation/Augmented Data',
                          save_format='jpeg'):
  i += 1
  if i>50:
    break
```
