# Yeni Bölüm

```python
from keras import models
from keras import layers
```

**Listing 2.4 The network architecture**


```python
network = models.Sequential()
```

```python
network.add(layers.Dense(512, activation='relu', input_shape=(28*28,)))
```

```python
network.add(layers.Dense(10, activation='softmax'))
```

**Listing 2.5 The compilation step**

```python
network.compile(optimizer='rmsprop',
                 loss='categorical_crossentropy',
                 metrics=['accuracy'])
```

**Listing 2.6 Preparing the image data**
Eğitimden önce, verilerimizi ağın beklediği şekle yeniden şekillendirerek ve tüm değerleri [0, 1] aralığında olacak şekilde ölçeklendirerek önceden işleyeceğiz. Önceden, örneğin eğitim görüntülerimiz, [0, 255] aralığında değerlerle uint8 tipi bir dizi şeklinde (60000, 28, 28) depolanıyordu. 0 ile 1 arasında değerlerle bir float32 şekil dizisine (60000, 28 * 28) dönüştürüyoruz.

```python
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255
```

```python
test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255
```

**Listing 2.7 Preparing the labels**
Ayrıca, 3. bölümde açıkladığımız bir adım olan etiketleri kategorik olarak kodlamamız gerekiyor:

```python
from keras.utils import to_categorical
```

```python
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

**Listing 2.8 Training the network**
Artık Keras'ta ağın "fit" yöntemine çağrı yoluyla yapılan ağımızı eğitmeye hazırız: modeli eğitim verilerine "fit".

```python
network.fit(train_images, train_labels, epochs=5, batch_size=128)
```

**Listing 2.9 Evaluating the network**


```python
test_loss, test_acc = network.evaluate(test_images, test_labels)
print('test_acc:', test_acc)
```
